```python
import pandas as pd
import numpy as np

indicators_df = pd.read_csv("Indicators.csv")
indicators_df.head()
```

```python
indicators_sm_df = indicators_df.loc[indicators_df["IndicatorName"]=="GDP growth (annual %)"]
indicators_sm_df.head()
```

```python
indicator_year=indicators_sm_df.loc[indicators_sm_df["Year"]==2014]
indicator_year
```

```python
indicator_year.set_index('CountryName', inplace=True)
indicator_year.rename(index=str, columns={'Country': 'CountryName', 'Value': 'Value'})
indicator_year.head()
```

```python
suicide_df = pd.read_csv('suicide-rates-1985-2016.csv')
suicide_df.head()
```

```python
suicide_df_year = suicide_df.loc[suicide_df["year"]==2014]
suicide_df_year.head()
```

```python
suicide_country=suicide_df_year.groupby(['country']).agg({'suicides/100k pop': 'sum'})
suicide_country.head()
```

```python
df3 = pd.merge(suicide_country, indicator_year, left_index=True, right_index=True)
df3.head()
```

```python
#Most suicides
desc=df3.sort_values(by=["Value"], ascending=True)
desc_new=desc.nlargest(5,"suicides/100k pop")
desc_new
```

```python
#least suicides
asce=df3.sort_values(by=["Value"], ascending=False)
asce_new=asce.nsmallest(5,"suicides/100k pop")
asce_new.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
 
# create data
x = df3["Value"]
y = df3["suicides/100k pop"]

 
# use the scatter function
plt.title("Suicide per GDP")
plt.xlabel("GDP")
plt.ylabel("Suicides per 100k")
plt.scatter(x, y)
plt.xcorr(x, y)
plt.show()


```

```python
df3["Value"].corr(df3["suicides/100k pop"])
```

```python
indicators_unemp = indicators_df.loc[indicators_df["IndicatorName"]=="Unemployment, total (% of total labor force)"]
indicators_unemp.head()
```

```python
unemp_year=indicators_unemp.loc[indicators_unemp["Year"]==2014]
unemp_year.head()
```

```python
unemp_year.set_index('CountryName', inplace=True)
unemp_year.rename(index=str, columns={'Country': 'CountryName', 'Value': 'Value'})
unemp_year.head()
```

```python
df_unemp = pd.merge(unemp_year, suicide_country, left_index=True, right_index=True)
df_unemp.head()
```

```python
# create data
x = df_unemp["Value"]
y = df_unemp["suicides/100k pop"]

 
# use the scatter function
plt.title("Suicide per Unemployment")
plt.xlabel("Unemployment %")
plt.ylabel("Suicides per 100k")
plt.scatter(x, y)
plt.xcorr(x, y)
plt.show()
```

```python
df_unemp["Value"].corr(df_unemp["suicides/100k pop"])
```

```python
suicide_df_11 = suicide_df.loc[suicide_df["year"]==2011]
suicide_country_11=suicide_df_11.groupby(['country']).agg({'suicides/100k pop': 'sum'})



indicators_gas = indicators_df.loc[indicators_df["IndicatorName"]=="CO2 emissions (kt)"]
gas_year=indicators_gas.loc[indicators_gas["Year"]==2011]
gas_year.set_index('CountryName', inplace=True)
gas_year.rename(index=str, columns={'Country': 'CountryName', 'Value': 'Value'})
# gas_year_red=indicators_gas.loc[indicators_gas["Value"]]
df_gas = pd.merge(gas_year, suicide_country_11, left_index=True, right_index=True)
df_gas.head()
```

```python
# create data
x = df_gas["Value"]
y = df_gas["suicides/100k pop"]

 
# use the scatter function
plt.title("Suicide per Total greenhouse gas emissions")
plt.xlim(0,300000)
plt.xlabel("Gas emissions")
plt.ylabel("Suicides per 100k")
plt.scatter(x, y)
plt.xcorr(x, y)
plt.show()
```
